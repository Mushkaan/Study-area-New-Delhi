```python
import ee
import geemap
```

```python
ee.Authenticate()
ee.Initialize(project= 'ee-mushkaan2024')
```

```python
m = geemap.Map()
m
```

```python
India= ee.FeatureCollection('projects/ee-mushkaan2024/assets/National_Boundary')
m.addLayer(India,{},'India')
m
```

```python
region= ee.Geometry.Polygon(
    [[[67.060547, 4.740675],
     [67.060547, 38.065392],
     [98.964844, 38.065392],
     [98.964844, 4.740675],
     [67.060547, 4.740675]]],
    None,
    False)
```

```python
ndvi= ee.ImageCollection("MODIS/061/MOD13A2").select('NDVI')
m
```

```python
ndvi_india = ndvi.map(lambda img: img.set('doy',
ee.Date(img.get('system:time_start')).getRelative('day', 'year')));
```

```python
distinctDOY = ndvi_india.filterDate('2000-02-18','2001-02-18');
filter= ee.Filter.equals(leftField='doy',rightField='doy')
join= ee.Join.saveAll('doy_matches')
joined= ee.ImageCollection(join.apply(distinctDOY,ndvi_india,filter))
```

```python
comp= joined.map(lambda img: ee.ImageCollection.fromImages(img.get('doy_matches')).reduce(ee.Reducer.mean()).set('doy',ee.Number(img.get('doy'))))
```

```python
viz_params = {
  'min': 0.0,
  'max': 9000.0,
  'palette': [
    'FFFFFF', 'CE7E45', 'DF923D', 'F1B555', 'FCD163', '99B718', '74A901',
    '66A000', '529400', '3E8601', '207401', '056201', '004C00', '023B01',
    '012E01', '011D01', '011301'
  ],
};
```

```python
rgbVis= comp.map(lambda img: img.visualize(bands=['NDVI_mean'],**viz_params).clip(India))
```

```python
gifParams = {
  'region': region,
  'dimensions': 600,
  'crs': 'EPSG:4326',
  'framesPerSecond': 10,
  'format':'gif'

}
```

```python
print(rgbVis.getVideoThumbURL(gifParams))
```
