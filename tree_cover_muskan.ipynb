```python
import geemap,geemap as map
import ee
ee.Authenticate()
ee.Initialize(project= 'ee-mushkaan2024')
```

```python
m= geemap.Map()
m
```

```python
ROI = ee.FeatureCollection("projects/ee-mushkaan2024/assets/indian_district")
m.addLayer(ROI,{},"ROI")

m
```

```python
India_dem = ee.Image("USGS/SRTMGL1_003").clip(ROI)
vis_map = {"min":0,"max":8700,'palette': ["#003300", "#006600",  "#009933",  "#33CC33", "#66FF66", "#E5FFCC",  "#CC9900",  "#996633",  "#804000",  "#662A00",  "#B2B2B2", "#D8D8D8", "#F0F0F0",  "#F5F5F5",  "#FFFFFF" ]}
m.addLayer(India_dem,vis_map,"India_dem")
m
```

```python
tree_cover = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2019-01-01","2019-12-31").select("Percent_Tree_Cover").mean().clip(ROI)
vis_tree = {"min":0,"max":78,'palette':['bbe029', '0a9501', '074b03']}
m.addLayer(tree_cover,vis_tree,"tree_cover")
m
```

```python
tree_cover1 = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2018-01-01","2018-12-31").select("Percent_Tree_Cover").mean().clip(ROI)
vis_tree1 = {"min":0,"max":85,'palette':['bbe029', '0a9501', '074b03']}
m.addLayer(tree_cover1,vis_tree1,"tree_cover1")
m
```

```python
tree_cover2 = ee.ImageCollection("MODIS/006/MOD44B").filterDate("2020-01-01","2020-12-31").select("Percent_Tree_Cover").mean().clip(ROI)
viz_tree2 = {"min":0,"max":78,'palette':['bbe029', '0a9501', '074b03']}
m.addLayer(tree_cover2,viz_tree2,"tree_cover2")
m
```
