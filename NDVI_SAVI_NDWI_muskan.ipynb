```python
map1=import_map.Map()
map2=import_map.Map()
map3=import_map.Map()
```

```python
study_area=ee.FeatureCollection('projects/ee-mushkaan2024/assets/kohima')
map1.addLayer(study_area,{},'Study_area')
map2.addLayer(study_area,{},'Study_area')
map3.addLayer(study_area,{},'Study_area')
map1.centerObject(study_area,9)
map2.centerObject(study_area,9)
map3.centerObject(study_area,9)
```

```python
landsat5_1985=ee.ImageCollection("LANDSAT/LT05/C02/T1").filterDate('1985-01-01','1985-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #1985
landsat5_1990=ee.ImageCollection("LANDSAT/LT05/C02/T1").filterDate('1990-01-01','1990-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #1990
landsat7_2000=ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2000-01-01','2000-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2000
landsat7_2010=ee.ImageCollection("LANDSAT/LE07/C02/T1").filterDate('2010-01-01','2010-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2010
landsat8_2020=ee.ImageCollection("LANDSAT/LC08/C02/T1").filterDate('2020-01-01','2020-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2020
landsat9_2024=ee.ImageCollection("LANDSAT/LC09/C02/T1").filterDate('2024-01-01','2024-12-31').filterMetadata('CLOUD_COVER','less_than',5).mean().clip(study_area) #2024
```

```python
ndvi_palette=(['#8B4513', '#9A5120', '#A9602D', '#B8703A', '#C58047', '#D49055', '#E3A063', '#E9B081', '#EEC09F', '#F3D0BD', '#F8E0DB', '#FCF0F9', '#FCF3E1', '#FDF6C8', '#FDF9AF', '#FEFC96', '#F7F580', '#E7E86C', '#D7DC58', '#C7CF45', '#B7C332', '#A7B71F', '#97AA0D', '#889E0A', '#789209', '#698607', '#5A7A06', '#4B6E05', '#3C6203', '#2D5602', '#1F4A01', '#154000', '#133800', '#123000', '#112800', '#102000', '#101800', '#101000', '#101000', '#122200'])
```

```python
#NDVI from 1985 to 2024
ndvi_1985=landsat5_1985.normalizedDifference(['B4','B3'])
ndvi_1990=landsat5_1990.normalizedDifference(['B4','B3'])
ndvi_2000=landsat7_2000.normalizedDifference(['B4','B3'])
ndvi_2010=landsat7_2010.normalizedDifference(['B4','B3'])
ndvi_2020=landsat8_2020.normalizedDifference(['B5','B4'])
ndvi_2024=landsat9_2024.normalizedDifference(['B5','B4'])
```

```python
ndwi_palette=(['#CEFFFD', '#76E3DE', '#76CCE3', '#40B4D3', '#4078D3', '#2126BB'])
```

```python
#NDWI from 1985 to 2024
ndwi_1985=landsat5_1985.normalizedDifference(['B2','B4'])
ndwi_1990=landsat5_1990.normalizedDifference(['B2','B4'])
ndwi_2000=landsat7_2000.normalizedDifference(['B2','B4'])
ndwi_2010=landsat7_2010.normalizedDifference(['B2','B4'])
ndwi_2020=landsat8_2020.normalizedDifference(['B3','B5'])
ndwi_2024=landsat9_2024.normalizedDifference(['B3','B5'])
```

```python
savi_palette=(['#FFFFFF','#D9B47C','#A78C59','#6BA96D','#2E8B57','#006400'])
```

```python
savi_1985 = landsat5_1985.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat5_1985.select('B4'),'RED': landsat5_1985.select('B3')})
savi_1990= landsat5_1990.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat5_1990.select('B4'),'RED': landsat5_1990.select('B3')})
savi_2000= landsat7_2000.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2000.select('B4'),'RED': landsat7_2000.select('B3')})
savi_2010= landsat7_2010.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat7_2010.select('B4'),'RED': landsat7_2000.select('B3')})
savi_2020= landsat8_2020.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat8_2020.select('B5'),'RED': landsat8_2020.select('B4')})
savi_2024= landsat9_2024.expression('((1.5) * (NIR - RED)) / (NIR + RED + 0.5)',{'NIR': landsat9_2024.select('B5'),'RED': landsat9_2024.select('B4')})
```

```python
from re import M
#ndvi add layer
map1.addLayer(ndvi_1985,{'min':-1,'max':1,'palette':ndvi_palette},'NDVI 1985')
map1.addLayer(ndvi_1990,{'min':-0.2,'max':0.5,'palette':ndvi_palette},'NDVI 1990')
map1.addLayer(ndvi_2000,{'min':-0.4,'max':0.5,'palette':ndvi_palette},'NDVI 2000')
map1.addLayer(ndvi_2010,{'min':-0.4,'max':0.3,'palette':ndvi_palette},'NDVI 2010')
map1.addLayer(ndvi_2020,{'min':-0.1,'max':0.4,'palette':ndvi_palette},'NDVI 2020')
map1.addLayer(ndvi_2024,{'min':0,'max':0.4,'palette':ndvi_palette},'NDVI 2024')
map1.addLayerControl()
map1
```
